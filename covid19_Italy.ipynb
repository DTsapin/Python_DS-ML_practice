```python
import matplotlib.ticker as ticker
plt.figure(figsize=(10,5))
plt.plot(data_lombardia["data"], data_lombardia["nuovi_positivi"], label='   COVID-19')
plt.xlabel("")
plt.ylabel("  COVID-19")
plt.yticks(np.arange(min(data_lombardia["nuovi_positivi"]), max(data_lombardia["nuovi_positivi"])+1, 4000.0))
plt.xticks([0, 685, 1318])
plt.subplots_adjust(bottom=0.15)
plt.tight_layout()
plt.legend()
```
